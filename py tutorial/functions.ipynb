```python
def farenheit(temp):
    return (temp - 32) * 5 / 9

print(farenheit(10))
```

```python
print(farenheit(10))
print(farenheit(20))
```

```python
def multia(*kids):
    print(type(kids))
    print(kids)
    print(kids[1])

multia('hello', 'arha')
```

```python
def keyarg(**kids):
    print(type(kids))
    for i in kids.keys():
        print(kids.get(i))

keyarg(name= 'arha', age = '2')
```
