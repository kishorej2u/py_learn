```python
def changecase(func):
  def myinner():
    return func().upper()
  return myinner

@changecase
def myfunction():
  return "Hello Sally"
@changecase
def my2():
  return "hola"

print(myfunction())
print(my2())
```

```python
def mult(nu):
    print(type(nu))
    def wrapper():
        print('before fun execution')
        out = nu()
        print('after fun execution')
        return out * 3
    return wrapper

@mult
def use1():
    a = 3 
    print('this is ')
    return a

@mult
def use2():
    b = 50
    print('this is 2nd input')
    return b
b = 5
print(use1())
print(use2())
```
